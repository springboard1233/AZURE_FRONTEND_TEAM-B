```python
import pandas as pd
import os

# Folder path where your CSV files are stored
folder_path = r"C:\Users\susha\OneDrive\Desktop\Azure Based Demand Forecasting & Capacity Optimization System\csv file"

# Load the CSV files
azure_data = pd.read_csv(os.path.join(folder_path, "azure_usage.csv"))
external_data = pd.read_csv(os.path.join(folder_path, "external_factors.csv"))

# ✅ Now check for missing values
print("Missing Values in Azure Data:")
print(azure_data.isnull().sum(), "\n")

print("Missing Values in External Data:")
print(external_data.isnull().sum(), "\n")

# ✅ Check for duplicate rows
print("Duplicate Rows in Azure Data:", azure_data.duplicated().sum())
print("Duplicate Rows in External Data:", external_data.duplicated().sum())

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV files
folder_path = r"../csv file"  # path to your CSV folder
azure_data = pd.read_csv(f"{folder_path}/azure_usage.csv")
external_data = pd.read_csv(f"{folder_path}/external_factors.csv.csv")

# Check for missing values
print("Missing Values in Azure Data:")
print(azure_data.isnull().sum(), "\n")

print("Missing Values in External Data:")
print(external_data.isnull().sum(), "\n")

# Check for duplicate rows
print("Duplicate Rows in Azure Data:", azure_data.duplicated().sum())
print("Duplicate Rows in External Data:", external_data.duplicated().sum())

# Plotting

# Histogram of numeric features in Azure Data
azure_data.hist(figsize=(10, 6), bins=20)
plt.suptitle("Distribution of Azure Usage Metrics")
plt.show()

# Line plot of cloud market demand over time
plt.figure(figsize=(12, 5))
sns.lineplot(data=external_data, x='date', y='cloud_market_demand')
plt.xticks(rotation=45)
plt.title("Cloud Market Demand Over Time")
plt.show()
```
